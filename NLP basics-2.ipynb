```python
# import the spacy module
import spacy
import en_core_web_sm

# load the language library from the spacy module
nlp = en_core_web_sm.load()

# use the nlp language library to parse the string
doc = nlp(u'Sachin Tendulkar is considered one of the best batsman in World Cricket.\n'
          u'He will always be remembered for the wonderful innings he played at Sharjah.')
```

```python
# number of tokens in the string
len(doc)
```

```python
# number of sentences in the string
sentences = [sent for sent in doc.sents]

len(sentences)
```

```python
# print second sentence in the string
print(sentences[1].text)
```

```python
# for each token in string, print the raw token, parts of speech, details of the token, and lemmatization
for token in doc:
    print("raw token:")
    print(token.text)
    print("parts of speech:")
    print(token.pos_)
    print("lemmatization:")
    print(token.lemma_)
    print("\n")
```
